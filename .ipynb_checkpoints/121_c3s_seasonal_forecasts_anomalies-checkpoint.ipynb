# Seasonal forecast anomalies

### About

This notebook provides you a practical introduction how to calculate seasonal forecast anomalies with data from the [ECMWF SEAS5 model](https://confluence.ecmwf.int/display/CKB/Description+of+SEAS5+C3S+contribution). The workflow shows you how to compute a hindcast climate mean for the climate reference period 1993 to 2016 and this hindcast climate mean is used to compute real-time forecast anomalies. At the end, the the computed anomalies are compared with precomputed seasonal forecast anomalies offered by the Copernicus Climate Data Store.
The example showcases global real-time forecast anomalies for May 2021 and five leadtime months (until October 2021).

## <a id='hindcast_climate mean'></a>2. Calculate seasonal hindcast climate mean

The example below will show you how to calculate seasonal forecast anomalies of total precipitation for September 2021 and for five leadtime months (until February 2022). 
The term anomaly refers to the deviation of a value from the long-term average. A positive anomaly indicates that the average precipitation of this month was higher than the reference value, while a negative anomaly indicates that during this specific month, it rained less and it was drier than the reference value.

The first step is to calculate the hindcast climate mean for September, which is the longterm average of precipitation for seasonal forecasts with a start runtime in September and five lead time months (until February). The reference period for the seasonal hindcast data is from 1993 to 2016 and the hindcast climate mean represents the average conditions observed over this period. 

Seasonal forecast data are disseminated in the `GRIB` data format. You can use the Python library [xarray](http://xarray.pydata.org/en/stable/) and its function `open_dataset` to read GRIB files from ECMWF. To read a GRIB file, you have to specify the keyword argument `engine` and specify `'cfgrib'`. [cfgrib](https://github.com/ecmwf/cfgrib) is a Python interface to map GRIB files to the NetCDF Common Data Model using [ecCodes](https://github.com/ecmwf/eccodes).

The result is a `xarray.Dataset` object with five dimensions: `number`, `time`, `step`, `latitude` and `longitude`. 

```python
ds = xr.open_dataset('./ecmwf_seas5_1993-2016_05_hindcast_monthly_tp.grib', engine='cfgrib')
ds
```

<br>

Per default, xarray does not load the GRIB file in a structure useful for us. You can specify a custom data structure by identifying the keyword argument `backend_kwargs` and specifically highlighting the two dimensions that shall be used for the time dimensions; `forecastMonth` and `time`. When you do this, you see that the resulting `xarray.Dataset` has five dimensions:
* `number`: Number of ensembles (25)
* `forecastMonth`: leadtime months (6, including the startime run)
* `time`: number of years (24, from 1993 to 2016)
* `latitude`: latitudes (181 -> 1 deg resolution)
* `longitude`: longitudes (360 --> 1 deg resolution)

```python
ds_hindcast = xr.open_dataset('./ecmwf_seas5_1993-2016_05_hindcast_monthly_tp.grib', engine='cfgrib', backend_kwargs=dict(time_dims=('forecastMonth', 'time')))
ds_hindcast
```

<br>

The next step is to load the data variable `tprate` as `xarray.DataArray`. You can load a DataArray from a Dataset by specifying the name of the variable in square brackets `[]`. A xarray.DataArray provides you additional attributes of the variable. For example, you see that the precipitation is given in accumulations with the unit m/s. Since these conversion from precipitation rate in m/s to total precipitation for a month in m depends on the number of days per month, we will execute the conversion in a later step. You find more information about the conversion of accumulated variables [here](https://confluence.ecmwf.int/pages/viewpage.action?pageId=197702790).

```python
tprate_hindcast = ds_hindcast['tprate']
tprate_hindcast
```

Since we are interested in creating the `hindcast climate mean`, the next step is to create the average over the 25 ensemble members and the 24 start years. You can use the function `mean()` in order to average over one or more given dimensions. The dimension we want to create the longterm mean are `number` (25 ensemble members) and `time` (start time years from 1993 to 2016). The result is a `xarray.DataArray` with three dimensions: `forecastMonth`, `latitude` and `longitude`.

The DataArray `tprate_hindcast_mean` is the hindcast climate mean for the month September and five leadtime months, over the reference period 1993 to 2016.

**Note:** with the averaging operation, your `xarray.DataArray` loses its attributes.

```python
tprate_hindcast_mean = tprate_hindcast.mean(['number', 'time'])
tprate_hindcast_mean
```

<br>

## <a id='seasonal_forecast_load'></a> 3. Load monthly seasonal forecast data and calculate seasonal forecast anomalies for May 2021

The next step is to load the real-time seasonal forecast data for May 2021 and five leadtime months. We open the dataset `ecmwf_seas5_2021_05_forecast_monthly_tp.grib` with the xarray function `open_dataset()`, where we again specify the two keyword arguments `engine` and `backend_kwargs`.

Once the the `xarray.Dataset` is loaded, you see that it has 4 dimensions:
* `number`: 51 ensemble members
* `forecastMonth`: 6 leadtime months, including May, the month of the model run
* `latitude`: latitude values
* `longitude`: longitude values

Compared to the hindcast data, we do not have the dimension `time`, as the data only holds information for the year 2021. Another difference is that seasonal forecast real-time data have 51 ensemble members, compared to hindcasts, which only have 25 ensemble members.

```python
seas5_forecast = xr.open_dataset('./ecmwf_seas5_2021_05_forecast_monthly_tp.grib', engine='cfgrib', backend_kwargs=dict(time_dims=('forecastMonth', 'time')))
seas5_forecast
```

Now, we can compute the seasonal forecast anomalies for May 2021. You can compute the anomalies by subtracting the longterm average (`tprate_hindcast_mean`) from the seasonal forecast real-time data for May 2021 (`seas5_forecast`).

The resulting `xarray.DataArray` has the anomaly information for each of the leadtime months from May 2021 to October 2021 relative to the reference period May 1993 to 2016.

```python
seas5_anomalies_202105 = seas5_forecast['tprate'] - tprate_hindcast_mean
seas5_anomalies_202105
```

<br>

A closer look at the coordinate information of `forecastMonth` shows you that it only gives integer values from 1 to 6, but not the actual time for which the forecast is valid. Let us assign a new coordinate with the name `valid_time` based on the dimension `forecastMonth`. This coordinate information shall provide us for each of the forecast months the valid timestamp. You can use the value of the time dimension as start date. The function `relativedelta()` lets you add a specific number of months to the start date.

In a second step, with the xarray function `assign_coords()`, we assign the `DateTimeIndex` object as new coordinate information.

```python
valid_time = [pd.to_datetime(seas5_anomalies_202105.time.values) + relativedelta(months=fcmonth-1) for fcmonth in seas5_anomalies_202105.forecastMonth]
seas5_anomalies_202105 = seas5_anomalies_202105.assign_coords(valid_time=('forecastMonth',valid_time))
seas5_anomalies_202105
```

Before visualizing the seasonal forecast anomalies, let us convert the precipitation accumulations in m/s to total precipitation for a month in mm. The conversion has to be done per month, as it is dependent on the number of days of a specific month. For this, we create a new dimension called `numdays`. The function `monthrange()` from the calendar library returns the weekday of first day of the month and number of days in a month, based on a given year and month. With the help of the new coordinate information `valid_time`, we create a list of number of days based on the valid time month. Again, with the function `assign_coords()`, we assign this list as new coordinate information.

```python
numdays = [monthrange(dd.year,dd.month)[1] for dd in valid_time]
seas5_anomalies_202105 = seas5_anomalies_202105.assign_coords(numdays=('forecastMonth',numdays))
seas5_anomalies_202105
```

We can now use the newly created coordinate information and convert for each month the precipitation accumulations to total precipitation. For this, we multiply the precipitation values with the number of days and then with (24 * 60 * 60) to retrieve precipitation values in m. As a last step, you can convert the values in m to precipitation in mm by multiplying the values with 1000.

```python
seas5_anomalies_202105_tp = seas5_anomalies_202105 * seas5_anomalies_202105.numdays * 24 * 60 * 60 * 1000
seas5_anomalies_202105_tp
```

<br>

As a last step before visualizing the data, we want to add the attributes `units` and `long_name` and update them, as they have changed with the previous workflow steps. You can simply specify the name of a attribute (e.g. `units`) and assign it a new value.

```python
seas5_anomalies_202105_tp.attrs['units'] = 'mm'
seas5_anomalies_202105_tp.attrs['long_name'] = 'Total precipitation anomaly'
seas5_anomalies_202105_tp
```

## <a id='visualize_seasonal'></a>4. Visualize seasonal forecast anomalies for a geographical subregion

Let us now visualize the seasonal forecast anomalies for a geographical subregion in two different ways: (i) as a `FacetGrid` plot with all ensemble members for one lead time and (ii) as a seasonal forecast anomaly plot. First, we create a geographical subset for South Asia with a bounding box of `[N:30, W:70, S:5, E:90]`. We define a function called `ds_latlon_subet()` which generated a geographical subset of a xarray data array. The latitude and longitude values that are outside the defined area are dropped.

```python
def ds_latlon_subset(ds,area,latname='latitude',lonname='longitude'):
    mask = (ds[latname]<=area[0]) & (ds[latname]>=area[2]) & (ds[lonname]<=area[3]%360) & (ds[lonname]>=area[1]%360)
    dsout = ds.where(mask,drop=True)
    return dsout
```

Now, we apply the function `ds_latlon_subset()` to the data array of the seasonal forecast anomalies. The subsetted array has 26 latitude grid points and 21 longitude grid points.

```python
subselregion = (30, 70, 5, 90) # North/West/South/East
seas5_anomalies_202105_SAsia = ds_latlon_subset(seas5_anomalies_202105_tp, subselregion)
seas5_anomalies_202105_SAsia
```

### <a id='facetgrid'></a> 4.1 FacetGrid plot visualization

Now, we can create a `FacetGrid` plot, which shows for one lead time the total precipitation anomaly of the control run and the 50 ensemble members. The plotting code below can be split in four main sections:
* **Define leadtime of seasonal forecast**: Define the leadtime of the seasonal forecast for which the FacetGrid plot shall visualize the ensemble members
* **FacetGrid plotting function**:: xarray built-in plotting function to create FacetGrids. With kwarg `col` you define the dimension that shall be used for the subplots and with `col_wrap` you define the number of columns of the FacetGrid.
* **Set a title for each subplot**: Loop over each subplot and set a customized title.
* **Add a superior plot**: Add a superior title describing the entire FacetGrid plot with the function `fig.suptitle()`.

```python
# Define leadtime of seasonal forecast
lead_time = 1

# FacetGrid plotting function
facet_plot = seas5_anomalies_202105_SAsia[:,lead_time,:,:].plot(
    col='number',
    col_wrap=10,
    cmap='bwr_r',
    cbar_kwargs={
        "orientation": "horizontal",
        "shrink": 0.3,
        "pad": 0.05,
        "label": seas5_anomalies_202105_SAsia.units
    },
    robust=True,
    vmin=-350,
    vmax=350)

# Loop over each subplot and set a subplot title
for i, ax in enumerate(facet_plot.axes.flat):
    if (i==0):
        ax.set_title('Control')
    else:
        ax.set_title("Forecast %d" % i)

# Add a superior title describing the FacetGrid plot
facet_plot.fig.suptitle('Total precipitation anomaly forecast in South Asia for ' + str(seas5_anomalies_202105_SAsia.valid_time[lead_time].data)[0:7] + ' - Control run and 50 ensemble members', fontsize=20, x=0.5, y=1.02)
```

### <a id='average_plot'></a>4.2 Plot of total precipitation anomalies for each seasonal forecast month

In a next step, we want visualize the total precipitation anomalies for each seasonal forecast month averaged over the subregion and compare it with the anomaly of the hindcast climatology.

The first step is to create the weighted average of the subregion defined above. For this, we first estimate the cell area with the cosine of the latitude. These weights are then applied when the data array is averaged over the two dimensions `latitude` and `longitude`. You can use the xarray function `weighted()` together with the reduced function `mean()` to create a weighted average of the geographical region. The result is a data array with two dimensions left: `number` and `forecastMonth`.

```python
weights =np.cos(np.deg2rad(seas5_anomalies_202105_SAsia.latitude))

anoms_SAsia = seas5_anomalies_202105_SAsia.weighted(weights).mean(['latitude','longitude'])
anoms_SAsia
```

The next step is to convert the data array into a pandas.Dataframe object with the function `to_dataframe()`. You might want to drop coordinates that are not needed. You can do this with the function `drop_vars()`.

```python
anoms_SAsia_df = anoms_SAsia.drop_vars(['time','surface','numdays']).to_dataframe(name='anomaly')
anoms_SAsia_df
```

Above, you see that the data frame is not a format that is practical for visualizing the the ensemble member anomalies for each leadtime month. For this reason, the next step is to bring the data frame with six row entries (for each valid time one) and 51 columns, representing the total precipitation anomalies of each ensemble member. The code below combines three pandas functions:
* **reset_index().drop()**: drops column forecastMonth, which will not be required in the final data frame
* **set_index()**: allows to define which column(s) shall be used as data frame index
* **unstack()**: transposes the data frame

```python
anoms_SAsia_df = anoms_SAsia_df.reset_index().drop('forecastMonth',axis=1).set_index(['valid_time','number']).unstack()
anoms_SAsia_df
```

The data frame above provides us for each ensemble member and forecast month the total precipitation anomaly. It is always helpful to compare the seasonal forecast anomalies for a specific month with the anomalies of the climatology of the hindcast data. This gives us an indication of the intensity of the anomaly.
For this reason, we come back to the data array of the seasonal forecast hindcast data (`tprate_hindcast`), which we loaded right at the beginning. Let us take this data array and also create a geographical subset with the function `ds_latlon_subset()`.

```python
tprate_hindcast_SAsia = ds_latlon_subset(tprate_hindcast, subselregion)
tprate_hindcast_SAsia
```

In a next step, we want to create the weighted average of the two dimensions `latitude` and `longitude`, to compute the averaged anomalies for the whole geographical subregion. We can use again the data array `weights` and apply those during the averaging process with the functions `weighted()` and `mean()`. The resulting array has three dimensions, `number`, `forecastMonth` and `time`.

```python
tprate_hindcast_SAsia = tprate_hindcast_SAsia.weighted(weights).mean(['latitude','longitude'])
tprate_hindcast_SAsia
```

The next step is then to compute the hindcast anomalies based on the hindcast climatology. You can create the hindcast seasonal forecast climatology by computing the mean over the two dimensions `number` and `time`. The result is a data array with six entries, representing the seasonal forecast climatology for each of the forecast months. In order to compute the anomalies,you subtract the climatology from the hindcast seasonal forecasts.

```python
anom_hindcast = tprate_hindcast_SAsia - tprate_hindcast_SAsia.mean(['number','time'])
anom_hindcast
```

Now, we also want to convert the hindcast anomalies from precipitation accumulations in m/s to total precipitation in m. The conversion has to be done per month, as it is dependent on the number of days of a specific month. For this, we create again a new dimension, called `numdays` The function `monthrange()` from the calendar library returns the weekday of first day of the month and number of days in a month, based on a given year and month. With the help of the new coordinate information `valid_time`, we create a list of number of days based on the valid time month. Again, with the function `assign_coords()`, we assign this list as new coordinate information.

```python
valid_time = [ pd.to_datetime(anom_hindcast.time.values[0]) + relativedelta(months=fcmonth-1) for fcmonth in anom_hindcast.forecastMonth]
numdays = [monthrange(dd.year,dd.month)[1] for dd in valid_time]
anom_hindcast = anom_hindcast.assign_coords(numdays=('forecastMonth',numdays))
anom_hindcast
```

We can now use the newly created coordinate information and convert for each month the precipitation accumulations to total precipitation. For this, we multiply the precipitation values with the number of days and then with (24 * 60 * 60) to retrieve precipitation values in m. As a last step, you can convert the values in m to precipitation in mm by multiplying the values with 1000.

```python
anom_hindcast_tp = anom_hindcast * anom_hindcast.numdays * 24 * 60 * 60 * 1000
anom_hindcast_tp
```

Now, based on the hindcast anomalies, for each of the forecast months, we can compute the minimum and maximum as well as the 33rd and 66th percentile. 

```python
P0 = anom_hindcast_tp.min(['number','time'])
P33 = anom_hindcast_tp.quantile(0.33,['number','time'])
P66 = anom_hindcast_tp.quantile(0.66,['number','time'])
P100 = anom_hindcast_tp.max(['number','time'])
```

<br>

The last step is to visualize the hindcast climatology boundaries as filled area and the enemble seasonal forecast anomalies for each forecast month. The visualisation code below has three main sections:
* **Initiate the figure**: Initiate a matplotlib figure
* **Plot the seasonal forecast anomalies + climatology boundaries as filled areas**: Plot the ensemble seasonal forecast anomalies as black circles and the hindcast climatology as filled areas 
* **Customise plot settings**: Add additonal features, such as title, x- and y-axis labels etc.

```python
# Initiate the figure
fig=plt.figure(figsize=(20, 10))
thisax=fig.gca()

# Plot the seasonal forecast anomalies + climatology boundaries as filled areas
thisax.plot(anoms_SAsia_df, marker='o', linestyle='', color='black', alpha=0.5, label='_nolegend_')
thisax.fill_between(anoms_SAsia_df.index,P66,P100,color='#5ab4ac',label='>P66')
thisax.fill_between(anoms_SAsia_df.index,P33,P66,color='lightgray',label='P33-P66')
thisax.fill_between(anoms_SAsia_df.index,P0,P33,color='#d8b365',label='<P33')

# Customize plot settings
plt.title('Total precipitation anomalies of seasonal forecasts for all members compared to the boundaries of the hindcast anomaly climatology\n', fontsize=14)
thisax.legend(fontsize=14,loc=1)
thisax.set_ylabel('mm\n', fontsize=12)
thisax.set_xlabel('\n Forecast month', fontsize=12)
```

## <a id='seasonal_3month'></a> 5. Calculate seasonal forecast 3-month anomalies

The last section aims to compute and visualize seasonal forecast 3-month anomalies. For this, the first step is to compute for the seasonal forecasts as well as the hindcast data three-month rolling averages. You can combine the xarray functions `rolling()` and `mean()` to do this.

```python
seas5_forecast_3m = seas5_forecast.rolling(forecastMonth=3).mean()
ds_hindcast_3m = ds_hindcast.rolling(forecastMonth=3).mean()

ds_hindcast_3m
```

The next step is then to calculate the 3-month seasonal forecast anomalies. For this, you first have to average the 3-month seasonal hindcasts over the dimensions `number` and `time`. Afterwards, you can subract the hindcast climatology from the 3-month seasonal forecasts.

```python
ds_hindcast_3m_hindcast_mean = ds_hindcast_3m.mean(['number','time'])
seas5_anomalies_3m_202105 = seas5_forecast_3m.tprate - ds_hindcast_3m_hindcast_mean.tprate
seas5_anomalies_3m_202105
```

Now, we want to compute the average of the 3-month seasonal forecast anomaly over each ensembel member. For this, we have to average over the dimension `number`. The final array has three dimensions, `forecastMonth`, `latitude` and `longitude`.

```python
seas5_anomalies_3m_202105_em = seas5_anomalies_3m_202105.mean('number')
seas5_anomalies_3m_202105_em
```

The last step before visualizing the 3-month seasonal forecast anomalies is again to convert the precipitation accumulations to total precipitation in mm. We repeat the same steps as we already did in previous steps:
* **1. Calculate number of days for each forecast month and add it as coordinate information to the data array**
* **2. Define names for the 3-month rolling archives, that give an indication over which months the average was built**
* **3. Convert the precipitation accumulations based on the number of days**
* **4. Add updated attributes to the data array**

```python
# Calculate number of days for each forecast month and add it as coordinate information to the data array
vt = [ pd.to_datetime(seas5_anomalies_3m_202105_em.time.values) + relativedelta(months=fcmonth-1) for fcmonth in seas5_anomalies_3m_202105_em.forecastMonth]
vts = [[thisvt+relativedelta(months=-mm) for mm in range(3)] for thisvt in vt]
numdays = [np.sum([monthrange(dd.year,dd.month)[1] for dd in d3]) for d3 in vts]
seas5_anomalies_3m_202105_em = seas5_anomalies_3m_202105_em.assign_coords(numdays=('forecastMonth',numdays))

# Define names for the 3-month rolling archives, that give an indication over which months the average was built
vts_names = ['{}{}{} {}'.format(d3[2].strftime('%b')[0],d3[1].strftime('%b')[0],d3[0].strftime('%b')[0], d3[0].strftime('%Y'))  for d3 in vts]
seas5_anomalies_3m_202105_em = seas5_anomalies_3m_202105_em.assign_coords(valid_time=('forecastMonth',vts_names))
seas5_anomalies_3m_202105_em

# Convert the precipitation accumulations based on the number of days
seas5_anomalies_3m_202105_em_tp = seas5_anomalies_3m_202105_em * seas5_anomalies_3m_202105_em.numdays * 24 * 60 * 60 * 1000

# Add updated attributes
seas5_anomalies_3m_202105_em_tp.attrs['units'] = 'mm'
seas5_anomalies_3m_202105_em_tp.attrs['long_name'] = 'Total precipitation ensemble mean anomaly'

seas5_anomalies_3m_202105_em_tp
```

Let us now visualize the 3-month total precipitation ensemble mean anomaly with the function `visualize_pcolormesh()` and `ipywidgets`. Widgets allow you to add interactive features to Jupyter notebooks and the feature we add is is a dropdown menue that offers the option to choose the three months periods for which the anomalies are averaged.

```python
def plot_leadt(ds, leadt_end):
    array_to_plot = ds.sel(forecastMonth=leadt_end)
    visualize_pcolormesh(data_array = array_to_plot, 
                     longitude = array_to_plot.longitude, 
                     latitude = array_to_plot.latitude, 
                     projection = ccrs.PlateCarree(), 
                     color_scale = 'bwr_r', 
                     unit = array_to_plot.units, 
                     long_name = array_to_plot.long_name+ '\n({})'.format(array_to_plot.valid_time.values), 
                     vmin= -350,
                     vmax = 350, 
                     set_global=True)

    return
    
dropdown_opts = [(vts_names[mm-1],mm) for mm in range(3,7)] 
a=widgets.interact(plot_leadt, ds=widgets.fixed(seas5_anomalies_3m_202105_em_tp), leadt_end=widgets.Dropdown(options=dropdown_opts,description='Valid Time:', style={'description_width': 'initial'}))
```
